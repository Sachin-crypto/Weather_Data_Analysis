```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Bangalore Weather Data - 1990 to 2022

```python
# Base Dataset Path
path = "D:/SACHIN/Desktop/Datasets/Temperature/"
```

```python
# Reading the df
bdf = pd.read_csv(path+"bangalore_temperature.csv", index_col=0, parse_dates=True)
bdf
```

```python
type(bdf.index)
```

```python
# Taking Data of 1990 and 2022
bdf_1990 = bdf.loc['1990']
bdf_2022 = bdf.loc['2022']
```

```python
bdf_1990
```

```python
bdf_2022.head()
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.plot(bdf_1990.index, bdf_1990['tmax'], marker='o', color='r')
plt.plot(bdf_1990.index, bdf_1990['tmin'], marker='o', color='m')
plt.plot(bdf_1990.index, bdf_1990['tavg'], marker='o', color='c')
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Legend
plt.legend(['Max Temperature', 'Min Temperature', 'Avg Temperature'])
# Title
plt.title("Bangalore Weather - 1990", fontsize='x-large')
# Padded layout
plt.tight_layout()
# Show
plt.show()
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.plot(bdf_2022.index, bdf_2022['tmax'], marker='o', color='r')
plt.plot(bdf_2022.index, bdf_2022['tmin'], marker='o', color='m')
plt.plot(bdf_2022.index, bdf_2022['tavg'], marker='o', color='c')
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Legend
plt.legend(['Max Temperature', 'Min Temperature', 'Avg Temperature'])
# Title
plt.title("Bangalore Weather - 2022", fontsize='x-large')
# Padded layout
plt.tight_layout()
# Show
plt.show()
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.plot(bdf_1990['tavg'], bdf_1990['prcp'], marker='o', color='r')
plt.plot(bdf_2022['tavg'], bdf_2022['prcp'], marker='o', color='c')
# Labels
plt.xlabel("Temperature")
plt.ylabel("Precipitation")
# Legend
plt.legend(['Avg Temperature(1990)','Avg Temperature(2022)'])
# Title
plt.title("Bangalore Weather - 1990 & 2022", fontsize='x-large')
# Padded layout
plt.tight_layout()
# Show
plt.show()
```

```python
# Count of values in year col
count_1990 = len(bdf_1990['tavg'].values)
count_2022 = len(bdf_2022['tavg'].values)
# Generating color based on count of values for scatter plot
colors_1990 = np.random.randint(count_1990, size=(count_1990))
colors_2022 = np.random.randint(count_2022, size=(count_2022))
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.scatter(bdf_1990.index, bdf_1990['tavg'], c=colors_1990, alpha=0.5, cmap='nipy_spectral')
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Title
plt.title("Bangalore Weather - 1990", fontsize='x-large')
# Colorbar
plt.colorbar()
# Show
plt.show()
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.scatter(bdf_2022.index, bdf_2022['tavg'], c=colors_2022, alpha=0.5, cmap='nipy_spectral')
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Title
plt.title("Bangalore Weather - 2022", fontsize='x-large')
# Colorbar
plt.colorbar()
# Show
plt.show()
```

```python
highest_ever = mdf.nlargest(n=10, columns='tmax')
highest_ever
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.bar(highest_ever.index, highest_ever['tmax'], color='r', width=100)
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Title
plt.title("Highest Mumbai Weather Recorded - 1990-2022", fontsize='x-large')
# Show
plt.show()
```

```python
lowest_ever = mdf.nsmallest(n=10, columns='tmax')
lowest_ever
```

```python
# Setting figure size
plt.figure(figsize=(16,8))
# Plotting data
plt.bar(lowest_ever.index, lowest_ever['tmax'], color='g', width=100)
# Labels
plt.ylabel("Temperature")
plt.xlabel("Day")
# Title
plt.title("Lowest Mumbai Weather Recorded - 1990-2022", fontsize='x-large')
# Show
plt.show()
```

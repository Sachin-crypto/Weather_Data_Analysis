```python
import pandas as pd
```

```python
dataset_base_path = "D:/SACHIN/Desktop/Datasets/Temperature/"
```

```python
df = pd.read_csv(dataset_base_path+"Bangalore.csv")
df
```

```python
# Converting time column into date type
df['time'] = pd.to_datetime(df['time'], format='mixed', dayfirst=True)
df
```

```python
# Setting time as index
df.set_index('time', inplace=True)
df
```

```python
# Remove index column header
df.index.name = None
df
```

```python
# Handling Missing Values - Filling values based on time
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = new_df.bfill()  # or new_df.fillna(method='bfill') deprecated
new_df
```

```python
new_df = new_df.round(1)
new_df
```

```python
new_df.to_csv(dataset_base_path+"bangalore_temperature.csv", index=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 8))
plt.ylabel('Date')
plt.xlabel('Temperature')
plt.title("Avg Temperature - Mumbai(1990-2022)", fontsize='large')
plt.plot(new_df.index, new_df['tavg'], marker='o', label='Avg Temperature', color='green')
# plt.plot(new_df.index, new_df['prcp'], marker='o', label='Precipitation (prcp)', color='red')
plt.plot(new_df.index, new_df['tmax'], marker='o', label='Max Temperature', color='red')
plt.plot(new_df.index, new_df['tmin'], marker='o', label='Min Temperature', color='yellow')
# plt.xticks(rotation=45)
plt.legend()
plt.tight_layout() # for padding
plt.show()
```
